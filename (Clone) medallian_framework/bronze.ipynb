```python
from pyspark.sql.functions import *
```

```python
volume_path = "/Volumes/dev_catalog/bronze/raw_file/shop_name_20251121.csv"
csv_df = spark.read.option("delimiter","|").csv(volume_path, header=True)
csv_df.display()
```

**Invalid**
- shop_id should filter row9:s, row12:null
- file_dt row4: invalid format

**Duplication**
- row5,6 : rowdup => dedup
- row7,8 : keydup => remove both

```python

# create control column
final_df =(
    csv_df
    .withColumn("_load_dt",current_date())
    .withColumn("_load_dttm",current_timestamp())
)
final_df.display()
```

```python
# load data to table (bronze mode)
final_df.write.mode("overwrite").saveAsTable("dev_catalog.bronze.shop_name_raw")
spark.sql("select * from dev_catalog.bronze.shop_name_raw").display()
```
